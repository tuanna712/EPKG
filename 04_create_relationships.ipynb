```python
import json
import pandas as pd
from neo4j import GraphDatabase

URI = "neo4j://localhost"
AUTH = ("neo4j", "167trungkinh")
```

### 1. Load data

```python
# Load file entities
file_path = 'data/processed/file_entities.csv'
files_data = pd.read_csv(file_path)
for i in range(len(files_data['Name'])):
    files_data['Name'][i] = files_data['Name'][i].replace("'", " ")
    files_data['ReportName'][i] = str(files_data['ReportName'][i]).replace("'", " ")
# Load Basin entities
file_path = 'data/processed/basin_entities.csv'
basin_data = pd.read_csv(file_path)
# Load Block entities
file_path = 'data/processed/block_entities.csv'
block_data = pd.read_csv(file_path)
# Load Well entities
file_path = 'data/processed/well_entities.csv'
well_data = pd.read_csv(file_path)
# Load Author entities
file_path = 'data/processed/author_entities.csv'
author_data = pd.read_csv(file_path)
# Load Contractor entities
file_path = 'data/processed/contractor_entities.csv'
contractor_data = pd.read_csv(file_path)
# Load TypeofFormat entities
file_path = 'data/processed/type_of_format_entities.csv'
type_of_format_data = pd.read_csv(file_path)
# Load TypesOfReport entities
file_path = 'data/processed/types_of_report_entities.csv'
types_of_report_data = pd.read_csv(file_path)
# Load StatusOfReport entities
file_path = 'data/processed/status_of_report_entities.csv'
status_of_report_data = pd.read_csv(file_path)
# Load Fields entities
file_path = 'data/processed/fields_entities.csv'
fields_data = pd.read_csv(file_path)
# Load Function entities
file_path = 'data/processed/function_entities.csv'
function_data = pd.read_csv(file_path)
# Load TypeOfFile entities
file_path = 'data/processed/type_of_file_entities.csv'
type_of_file_data = pd.read_csv(file_path)
```

### 2. Basin Relationships

```python
# Remove NaN values of Basin column
basin_data = basin_data.dropna(subset=['Basin'])

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    for i in range(len(basin_data)):
        query = """
        MATCH (s:File {{name: '{}'}}), (o:Basin {{name: 'SongHong'}})
        CREATE (s)-[:HAS_INFORMATION_OF]->(o)
        CREATE (o)-[:HAS_FILE]->(s)
        """.format(
                    basin_data['Name'][i],
                    )
        print(query)
        driver = GraphDatabase.driver(URI, auth=AUTH)
        driver.execute_query(query_=query, database_="neo4j")
```

### 3. Block Relationships

#### 3.1 Basin-Block Relationship

```python
# Remove NaN values of Basin column
block_data = block_data.dropna(subset=['Block'])
block_data = block_data[block_data.Block != 'All']
block_data_unique = block_data['Block'].unique()

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    for i in range(len(block_data_unique)):
        query = """
        MATCH (s:Block {{name: '{}'}}), (o:Basin {{name: 'SongHong'}})
        CREATE (s)-[:BELONG_TO]->(o)
        CREATE (o)-[:HAS_BLOCK]->(s)
        """.format(
                    block_data_unique[i],
                    )
        print(query)
        driver = GraphDatabase.driver(URI, auth=AUTH)
        driver.execute_query(query_=query, database_="neo4j")
```

#### 3.2 File-Block Relationship

```python
# Remove NaN values of Basin column
block_data = block_data.dropna(subset=['Block'])
block_data = block_data[block_data.Block != 'All']
block_data = block_data.reset_index(drop=True)

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    for i in range(len(block_data)):
        query = """
        MATCH (s:Block {{name: '{}'}}), (o:File {{name: '{}'}})
        CREATE (s)-[:HAS_FILE]->(o)
        CREATE (o)-[:HAS_INFORMATION_OF]->(s)
        """.format(
                    block_data['Block'][i],
                    block_data['Name'][i],
                    )
        print(query)
        driver = GraphDatabase.driver(URI, auth=AUTH)
        driver.execute_query(query_=query, database_="neo4j")
```

### 4. Well Relationships

```python
well_data = well_data.dropna(subset=['WellName'])
well_data = well_data[well_data.WellName != 'Well Completion Report']
well_data = well_data[well_data.WellName != '103-TH-1X and 103-TG-1X']
well_data = well_data.reset_index(drop=True)

well_block_data = well_data[['WellName', 'Block']]
well_block_data = well_block_data[well_data.Block != '[]']
well_block_data = well_block_data[well_data.Block != "['102-106', '103-107']"]
well_block_data = well_block_data[well_data.Block != "['118', '119']"]
well_block_data = well_block_data.dropna(subset=['Block'])
```

##### 4.1 Well-Block Relationship

```python
well_block_data['Block'] = well_block_data['Block'].str.replace(r"[\[\]']", '')
well_block_data.drop_duplicates(subset=['WellName', 'Block'], keep='first', inplace=True)
well_block_data = well_block_data.reset_index(drop=True)

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    for i in range(len(well_block_data)):
        query = """
        MATCH (s:Well {{name: '{}'}}), (o:Block {{name: '{}'}})
        CREATE (s)-[:BELONG_TO]->(o)
        CREATE (o)-[:HAS_WELL]->(s)
        """.format(
                    well_block_data['WellName'][i],
                    well_block_data['Block'][i],
                    )
        print(query)
        try:
            driver = GraphDatabase.driver(URI, auth=AUTH)
            driver.execute_query(query_=query, database_="neo4j")
        except Exception as e:
            print("Import to Graph got error: ",e)
```

##### 4.2 File-Well Relationship

```python
well_data = well_data[well_data.WellName != "119-CH-1X & 118-CVX-1X"]
well_data = well_data[well_data.WellName != 'MVHN02-CR-1X , MVHN02-DV-1X']
well_data = well_data.reset_index(drop=True)

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    for i in range(len(well_data)):
        query = """
        MATCH (s:File {{name: '{}'}}), (o:Well {{name: '{}'}})
        CREATE (s)-[:HAS_INFORMATION_OF]->(o)
        CREATE (o)-[:HAS_FILE]->(s)
        """.format(
                    well_data['Name'][i],
                    well_data['WellName'][i],
                    )
        print(query)
        driver = GraphDatabase.driver(URI, auth=AUTH)
        driver.execute_query(query_=query, database_="neo4j")
```
