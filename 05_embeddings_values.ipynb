### Embedding Function

```python
def get_embeddings(data:pd.DataFrame, column_name:str, saving_path:str):
    # Remove nan values of ReportAuthor
    data = data[data[column_name].notna()]
    unique_data = data[column_name].unique()
    embeded_unique_data = {}
    for i, item in enumerate(unique_data):
        print(f"{i+1}/{len(unique_data)}")
        emb = OAI.get_embedding(item)
        embeded_unique_data[item] = emb
    data[column_name + "Embeddings"] = data[column_name].apply(lambda x: embeded_unique_data[x])
    data = data.reset_index(drop=True)
    # Save to csv
    data.to_csv(saving_path, index=False)
    return data
```
